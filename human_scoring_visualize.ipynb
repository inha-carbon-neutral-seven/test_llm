```python
import statistics

def string_list2float_list(list):
    return [float(num) for num in list]
```

```python
average_acc_gpt4tnldcc_1n2 = df['average_acc_gpt4t&ldcc_1']
average_korean_gpt4tnldcc_1n2 = df['average_korean_gpt4t&ldcc_1']
average_acc_gpt35tnllama_1n2 = df['average_acc_gpt35t&llama_1']
average_korean_gpt35tnllama_1n2 = df['average_korean_gpt35t&llama_1']

simple_Q_gpt4t_Accuracy_avg =  string_list2float_list(list(average_acc_gpt4tnldcc_1n2[1:11]))
simple_Q_gpt3t_Accuracy_avg =  string_list2float_list(list(average_acc_gpt35tnllama_1n2[1:11]))
simple_Q_ldcc_Accuracy_avg =   string_list2float_list(list(average_acc_gpt4tnldcc_1n2[14:24]))
simple_Q_llama2_Accuracy_avg = string_list2float_list(list(average_acc_gpt35tnllama_1n2[14:24]))

print(simple_Q_gpt4t_Accuracy_avg)
print(simple_Q_gpt3t_Accuracy_avg)
print(simple_Q_ldcc_Accuracy_avg)
print(simple_Q_llama2_Accuracy_avg)


simple_Q_gpt4t_Korean_avg =  string_list2float_list(list(average_korean_gpt4tnldcc_1n2[1:11]))
simple_Q_gpt3t_Korean_avg =  string_list2float_list(list(average_korean_gpt35tnllama_1n2[1:11]))
simple_Q_ldcc_Korean_avg =   string_list2float_list(list(average_korean_gpt4tnldcc_1n2[14:24]))
simple_Q_llama2_Korean_avg = string_list2float_list(list(average_korean_gpt35tnllama_1n2[14:24]))

print(simple_Q_gpt4t_Korean_avg)
print(simple_Q_gpt3t_Korean_avg)
print(simple_Q_ldcc_Korean_avg)
print(simple_Q_llama2_Korean_avg)
```

```python
average_acc_gpt4tnldcc_1n2 = df['average_acc_gpt4t&ldcc_1']
average_korean_gpt4tnldcc_1n2 = df['average_korean_gpt4t&ldcc_1']
average_acc_gpt35tnllama_1n2 = df['average_acc_gpt35t&llama_1']
average_korean_gpt35tnllama_1n2 = df['average_korean_gpt35t&llama_1']

Advan_Q_gpt4t_Accuracy_avg = string_list2float_list(list(average_acc_gpt4tnldcc_1n2[27:32]))
Advan_Q_gpt3t_Accuracy_avg = string_list2float_list(list(average_acc_gpt35tnllama_1n2[27:32]))
Advan_Q_ldcc_Accuracy_avg = string_list2float_list(list(average_acc_gpt4tnldcc_1n2[35:40]))
Advan_Q_llama2_Accuracy_avg = string_list2float_list(list(average_acc_gpt35tnllama_1n2[35:40]))

print(Advan_Q_gpt4t_Accuracy_avg)
print(Advan_Q_gpt3t_Accuracy_avg)
print(Advan_Q_ldcc_Accuracy_avg)
print(Advan_Q_llama2_Accuracy_avg)


Advan_Q_gpt4t_Korean_avg = string_list2float_list(list(average_korean_gpt4tnldcc_1n2[27:32]))
Advan_Q_gpt3t_Korean_avg =  string_list2float_list(list(average_korean_gpt35tnllama_1n2[27:32]))
Advan_Q_ldcc_Korean_avg =   string_list2float_list(list(average_korean_gpt4tnldcc_1n2[35:40]))
Advan_Q_llama2_Korean_avg = string_list2float_list(list(average_korean_gpt35tnllama_1n2[35:40]))

print(Advan_Q_gpt4t_Korean_avg)
print(Advan_Q_gpt3t_Korean_avg)
print(Advan_Q_ldcc_Korean_avg)
print(Advan_Q_llama2_Korean_avg)
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Simple questions accuracy', 'Simple questions korean ability', 'Advanced question accuracy', 'Advanced question korean ability']

scores = {
    'ChatGPT (GPT3.5)': [statistics.mean(simple_Q_gpt3t_Accuracy_avg), statistics.mean(simple_Q_gpt3t_Korean_avg), statistics.mean(Advan_Q_gpt3t_Accuracy_avg), statistics.mean(Advan_Q_gpt3t_Korean_avg) ],
    'ChatGPT (GPT4-Turbo)': [statistics.mean(simple_Q_gpt4t_Accuracy_avg), statistics.mean(simple_Q_gpt4t_Korean_avg), statistics.mean(Advan_Q_gpt4t_Accuracy_avg), statistics.mean(Advan_Q_gpt4t_Korean_avg) ],
    'LDCC-SOLAR': [statistics.mean(simple_Q_ldcc_Accuracy_avg), statistics.mean(simple_Q_ldcc_Korean_avg), statistics.mean(Advan_Q_ldcc_Accuracy_avg), statistics.mean(Advan_Q_ldcc_Korean_avg) ],
    'Llama': [statistics.mean(simple_Q_llama2_Accuracy_avg), statistics.mean(simple_Q_llama2_Korean_avg), statistics.mean(Advan_Q_llama2_Accuracy_avg), statistics.mean(Advan_Q_llama2_Korean_avg) ],
}

n_categories = len(categories)

# X locations for the groups
ind = np.arange(n_categories)

# The width of the bars
bar_width = 0.1

# Create subplots
fig, ax = plt.subplots(figsize=(15, 8))

# Create bars for each model/group
for i, (model, model_scores) in enumerate(scores.items()):
    ax.bar(ind + i * bar_width, model_scores, bar_width, label=model)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Categories')
ax.set_ylabel('Average Score (%)')
ax.set_title('Mean test scores for each LLM by category')
ax.set_xticks(ind + bar_width * (len(scores) - 1) / 2)
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
